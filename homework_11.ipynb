### Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
ratings = pd.read_csv('C:\\Users\\2-hom\\Documents\\Data_analysis\\Pyton_Notebook\\ml-latest-small\\ratings.csv')
ratings.head(100)
```

```python
ratings_1 = ratings.groupby('movieId').mean('rating')
ratings_1[['rating']] = ratings_1[['rating']].astype('int')
ratings_1.head(10)
```

```python
movies = pd.read_csv('C:\\Users\\2-hom\\Documents\\Data_analysis\\Pyton_Notebook\\ml-latest-small\\movies.csv')
movies.head(10)
```

```python
joined = movies.merge(ratings_1, on='movieId', how='left')
movies_1 = joined.drop(columns=['userId', 'timestamp'])
# movies_1.head(10)
isNaN = movies_1[movies_1.isna().any(axis=1)]
isNaN
```

#### Вариант 1: 
принимает весь датафрейм - отрабатвает быстрее, чем если передавать в функцию построчно

```python
def raiting_class_simple(data):
    ls=[]
    for line in data.rating:
        if line >= 4.5:
            ls.append('высокий рейтинг')
        elif 2 < line < 4.5:
            ls.append('средний рейтинг')
        elif line <= 2:
            ls.append('низкий рейтинг')
        elif pd.isna(line) == True:
            ls.append('нет оценок')
    data['class'] = ls
    return data
    
movies_2 = raiting_class_simple(movies_1)
movies_2.head(10)
```

#### Вариант 2: 
передает в функцию построчно

```python
def raiting_class(row):
    if row.rating >= 4.5:
        return 'высокий рейтинг'
    elif 2 < row.rating < 4.5:
        return 'средний рейтинг'
    elif row.rating <= 2:
        return 'низкий рейтинг'
    elif pd.isna(row.rating) == True:
        return 'нет оценок'
        
movies_1['class'] = movies_1.apply(raiting_class, axis=1)
movies_1.head(10)
```

### Задание 2

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.

```python
data = pd.read_csv('C:\\Users\\2-hom\\Documents\\Data_analysis\\Pyton_Notebook\\Python_11_pandas\\keywords.csv', encoding='utf-8')
data.head()
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
```

```python
def geo_cl(row):
    k = 'undefined'
    for key, value in geo_data.items():
        for i in range(0,len(geo_data[key])-1):
            if geo_data[key][i] in row['keyword']:
                k = key
    return k
```

```python
data['region'] = data.apply(geo_cl, axis=1)
```
