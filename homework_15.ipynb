#### Графики зависимости уровня счасть 

```python
plt.plot(happiness['Уровень счастья'], happiness['ВВП на душу населения'])
plt.title('Зависимость уровня счастье от ВВП')
plt.xlabel('Уровень счастья')
plt.ylabel('ВВП на душу населения')
```

Видна прямая зависимость уровня счастья от размера ВВП на душу населения. Это объясняется высоким уровнем экономики, а следовательно ростом благосостояния. 

```python
plt.plot(happiness['Уровень счастья'], happiness['Социальная поддержка'])
plt.title('Зависимость уровня счастье от социальной поддержки')
plt.xlabel('Уровень счастья')
plt.ylabel('Социальная поддержка')
```

Зависимость уровня счастья от социальной поддержки населения также прямая, причем с ростом уровня соц.поддержки зависимость становится ближе к функциональной с незначительными выбросами

```python
plt.plot(happiness['Уровень счастья'], happiness['Продолжительность жизни'])
plt.title('Зависимость уровня счастье от продолжительности жизни')
plt.xlabel('Уровень счастья')
plt.ylabel('Продолжительность жизни')
```

Аналогично графикам выше наблюдается прямая зависимость уровня счастья от средней продолжительности жизни. 

```python
plt.plot(happiness['Уровень счастья'], happiness['Свобода выбора'])
plt.title('Зависимость уровня счастье от свобод')
plt.xlabel('Уровень счастья')
plt.ylabel('Свобода выбора')
```

Интересно, что при оценке зависимости уровня счастья от еаличия гражданских свобод выбросов значительно больше, чем при показателях социального багополучия населения: ВВП на душу населения, программы социальной поддержки, продолжительность жизни. По графику можно сделать гиотезу о том, что только при высоких значения показателей свобод уровень счастья находится в прямой зависимость.

```python
plt.plot(happiness['Уровень счастья'], happiness['Щедрость'])
plt.title('Зависимость уровня счастье от щедрости')
plt.xlabel('Уровень счастья')
plt.ylabel('Щедрость')
```

Интересно, что щедрость и счастье оказывают очень не значительное влияние друг на друга.

```python
plt.plot(happiness['Уровень счастья'], happiness['Уровень коррупции'])
plt.title('Зависимость уровня счастье от уровня коррупции')
plt.xlabel('Уровень счастья')
plt.ylabel('Уровень коррупции')
```

Также интересно, что уровень счастья практически не зависит от уровня коррупции.

#### Гистограмма топ-5  стран с максимальным и минимальным уровнем счастья

```python
top_5 = happiness.head(5).append(happiness.tail(5))
top_5 = top_5.drop(columns=['Рейтинг','Уровень счастья'], axis=1)
top_5
```

```python
top_5.plot(kind='bar', title='Топ 5 стран с максимальным и минимальным уровнем счастья',
                             xlabel='Значение показателей',
                             ylabel='Страны')
```

Гистограммы топ-5 стран с наибольши уровнем счастья и топ-5 стран с наименьшим уровнем счастья подтверждают выводы, сделанные выше по графикам: наибольшее влияние на уровень счастья оказывает уровень развития социальных программ, размер вВП на душу населения и продолжительность жизни; уровень щедрости находится примерно на одинаковом уровне как в самых счастливых странах, так и в странах-аутсайдерах. Интресно, что в странах-аутсайдерах коррупция практически не оценивается, можно сделать предположение, что коррупция зависит от благосостояния экономики.

#### Точечная диаграмма

```python
# посмотрим на взаимосвязь коррупции, социальной поддержки и ВВП

happiness.plot(kind='scatter', 
        x='Социальная поддержка', 
        y='Уровень коррупции', 
        title='Корреляция между коррупцией и социальной поддержкой')
plt.show()

happiness.plot(kind='scatter', 
        x='ВВП на душу населения', 
        y='Уровень коррупции', 
        title='Корреляция между коррупцией и ВВП на душу населений')
plt.show()

happiness.plot(kind='scatter', 
        x='Щедрость', 
        y='Уровень коррупции', 
        title='Корреляция между коррупцией и щедростью')
plt.show()

happiness.plot(kind='scatter', 
        x='Свобода выбора', 
        y='Уровень коррупции', 
        title='Корреляция между коррупцией и свободой выбора')
```

Действительно, наше предположение о связи коррупции с уровнем развития экономики подтверждается точечными диаграммами:

1) на первой диаграмме видна корреляция коррупции с развиитем социальных программ: при низком уровне финасирования социальных программ отсутсвует и коррупция, при высоком уровне социальных программ появляется и коррупция и идет ее рост.

2) на второй диаграмме виден рост коррупции с ростом ВВП

3) а вот после показателя щедрости 0,3 коррупция начинает пропадать, т.к. растет сознание людей.

4) странный на первй взгляд вывод дает четвертая диаграмма: страны с большим количеством прав и свобод оцениваются высоким уровнем коррупции и наоборот: чем ниже оуенены права и свободы, тем ниже и коррупция. Скорее всего низкий уровень коррупции в странах с низким уровнем прав и свобод связан с боязнью жестоких законов. 

####  Boxplot

```python
happiness['ВВП на душу населения'].plot(kind='box',
                   title='ВВП на душу населения', grid=True)
plt.show()

happiness['Социальная поддержка'].plot(kind='box',
                   title='Социальная поддержка', grid=True)
plt.show()

happiness['Продолжительность жизни'].plot(kind='box',
                   title='Продолжительность жизни', grid=True)
plt.show()

happiness['Свобода выбора'].plot(kind='box',
                   title='Свобода выбора', grid=True)
plt.show()

happiness['Щедрость'].plot(kind='box',
                   title='Щедрость', grid=True)
plt.show()

happiness['Уровень коррупции'].plot(kind='box',
                   title='Уровень коррупции', grid=True)
```

Количественные данные дают значительно меньше выбросов, качественные данные имеют значительные выбросы. ПОказатель уровня коррупции характеризуется самыми значительными выбросами.
