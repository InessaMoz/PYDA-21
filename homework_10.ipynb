### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

ratings = pd.read_csv('C:\\Users\\2-hom\\Documents\\Data_analysis\\Pyton_Notebook\\ml-latest-small\\ratings.csv')
ratings.head()
```

```python
movies = pd.read_csv('C:\\Users\\2-hom\\Documents\\Data_analysis\\Pyton_Notebook\\ml-latest-small\\movies.csv')
movies.head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

#### Вариант 1
подсчет уникального количества строк по значению столбца title

```python
joined[joined['rating']==5.0]['title'].value_counts().head(1)
```

#### Вариант 2
1. делаем выборку рейтингов с оценкой 5.0
2. группируем по ниаменованию фильмов
3. подсчитываем кол-во оценок, поставленных фильму
4. сортируем по убыванию, выводи 1 элемент списка

```python
joined[joined['rating']==5.0].groupby('title').count().sort_values('rating', ascending=False).head(1)
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
ds=pd.read_csv('C:\\Users\\2-hom\\Documents\\Data_analysis\\Pyton_Notebook\\power.csv')
ds.head()
```

```python
ds2 = ds[(ds['year'] >= 2005) & (ds['year'] <= 2010) & (ds['quantity'] > 0)]
ds3 = ds2[(ds2['category'] == 4) | (ds2['category'] == 12)| (ds2['category'] == 21)]  

summ = ds3[ds3['country'].str.contains('|'.join(['Lat','Lit','Est']))]['quantity'].sum()
print('Cуммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года:',summ)
```
