### Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
## поиск выброса через межквартильный размах для столбца 4:ректальная температура

q1 = horse['4.T_рек'].quantile(0.25)
q3 = horse['4.T_рек'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse[horse['4.T_рек'].between(lower_bound, upper_bound, inclusive=True)].sort_values('4.T_рек')
remove_outliers
```

```python
# что это за выброс?
pd.concat([horse, remove_outliers]).drop_duplicates(keep=False)
```

```python
## поиск выброса через межквартильный размах для столбца 5:пульс

q1 = horse['5.Пульс'].quantile(0.25)
q3 = horse['5.Пульс'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse[horse['5.Пульс'].between(lower_bound, upper_bound, inclusive=True)].sort_values('5.Пульс')
remove_outliers
```

```python
# что это за выброс?
pd.concat([horse, remove_outliers]).drop_duplicates(keep=False)
```

```python
## поиск выброса через межквартильный размах для столбца 16:назогастральный рефлюкс, PH

q1 = horse['16.Рефлюкс_PH'].quantile(0.25)
q3 = horse['16.Рефлюкс_PH'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse[horse['16.Рефлюкс_PH'].between(lower_bound, upper_bound, inclusive=True)].sort_values('16.Рефлюкс_PH')
remove_outliers
```

```python
# что это за выброс?
pd.concat([horse['16.Рефлюкс_PH'], remove_outliers]).drop_duplicates(keep=False)
```

```python
## поиск выброса через межквартильный размах для столбца 22:общий белок брюшной полости

q1 = horse['22.Об.белок_2'].quantile(0.25)
q3 = horse['22.Об.белок_2'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse[horse['22.Об.белок_2'].between(lower_bound, upper_bound, inclusive=True)].sort_values('22.Об.белок_2')
remove_outliers
```

```python
# что это за выброс?
pd.concat([horse, remove_outliers]).drop_duplicates(keep=False)
```
