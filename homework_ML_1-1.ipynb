#### 1. Построение логистической регрессии

```python
from sklearn.linear_model import LogisticRegression
```

```python
# выбитраем признаки
selectedColumns = data[ [ 'age', 'workclass', 'education', 'educational-num', 'marital-status', 'occupation', 'race', 'gender', 'hours-per-week', 'native-country', 'income' ] ]

# столбцы 'workclass', 'education', 'marital-status', 'occupation', 'race', 'gender', 'native-country' являются категориальными переменными
# переведем их в значения 0 и 1, добавив столбцы с соответствующими названиями
X = pd.get_dummies( selectedColumns, columns = [ 'workclass', 'education', 'marital-status', 'occupation', 'race', 'gender', 'native-country'] )

# столбец income является целевой переменной, удаляем его из X
del X['income']
X.head()
```

```python
# целевая переменная (столбец income) снова является категориальной
# переведем значения столбца в числа, оставив один столбец

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit( data['income'] )
```

```python
le.classes_
```

```python
# пример "расшировки" столбца income

le.transform( [ '<=50K', '>50K' ] )
```

```python
# записываем в переменную y преобразованный столбец income

y = pd.Series( data = le.transform( data['income'] ) )
y.head()
```

```python
from sklearn.pipeline import make_pipeline # используем пайплайны для удобства
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split 

model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))
```

```python
model = LogisticRegression() # берем в качестве модели логистическую регресиию из scikit-learn
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
X
```

```python
# обучаем на части датасета (train)

model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
predictions[:5] # посмотрим как выглядят наши предсказания
```

```python
model.predict(X_test) # получаем массив
```

```python
model.predict_proba(X_test)
```

```python
predictions[:5]
```

```python
y_pred = model.predict(X_test)   # прогнозные данные модели
y_pred
```

```python
# сравним пронозы, построенные моделью, с фактом
res_df = X_test.copy()

res_df['y_pred'] = y_pred
res_df['y_test'] = y_test
res_df
```

```python
# Оценим точность предсказания на обучающей и тестовой выборках
```

```python
model.score(X_train, y_train) # метод возвращает значение коэффициента детерминации - точность предсказания на обучающей выборке
```

```python
model.score(X_test, y_test) # метод возвращает значение коэффициента детерминации - точность предсказания на тестовой выборке
```

#### 2. Построение модели  SVM

```python
from sklearn.svm import SVC
# используем пайплайны
clf = make_pipeline(StandardScaler(), SVC(gamma='auto')) 
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train) 
```

```python
clf.score(X_test, y_test)
```

```python
svc = SVC(gamma='auto')
svc.fit(X_train, y_train)
```

```python
svc.score(X_train, y_train)
```

```python
svc.score(X_test, y_test)
```
