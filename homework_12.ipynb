### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

* если источник traffic_source равен yandex или google, то в source_type ставится organic

* для источников paid и email из России - ставим ad

* для источников paid и email не из России - ставим other

* все остальные варианты берем из traffic_source без изменений


```python
import pandas as pd
```

```python
log = pd.read_csv('C:\\Users\\2-hom\\Documents\\Data_analysis\\Pyton_Notebook\\Pandas_12_13\\visit_log.csv', sep=';')
log.head()
```

#### Вариант 1
используем метод loc

```python
log.loc[(log.traffic_source == 'yandex') | (log.traffic_source == 'google'), 'source_type'] = 'organic'
log.loc[((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region=='Russia'), 'source_type'] = 'ad'
log.loc[((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region!='Russia'), 'source_type'] = 'other'
log.loc[log.source_type.isnull(), 'source_type'] = log['traffic_source']
log.head()
```

#### Вариант 2
С помощью функции

```python
def filtered(row):
    if row['traffic_source']=='yandex':
        return 'organic'
    elif row['traffic_source']=='google':
        return 'organic'
    elif row['traffic_source']=='paid':
        if row['region']=='Russia':
            return 'ad'
        else:
            return 'other'
    elif row['traffic_source']=='email':
        if row['region']=='Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return row['traffic_source'] 
```

```python
log['source_type']=log.apply(filtered, axis=1)
log.head()
```

### Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('C:\\Users\\2-hom\\Documents\\Data_analysis\\Pyton_Notebook\\Pandas_12_13\\ml-latest-small\\ratings.csv')
ratings.head()
```

```python
df = ratings.groupby('userId').count().sort_values('movieId', ascending=False)
df = df.drop(columns = ['rating', 'timestamp'])
df.rename(columns={'movieId': 'count1'}, inplace=True)
df.head()
```

```python
ratings_grouped = ratings.groupby('userId').agg([min, max])
ratings_grouped.head()
```

```python
ratings_grouped['diff'] = (ratings_grouped['timestamp']['max'] - ratings_grouped['timestamp']['min'])
ratings_grouped.head()
```

```python
ratings100 = ratings_grouped.copy()
ratings100[df.columns] = df
ratings100 = ratings100.drop(ratings100[ratings100.count1 < 100].index)
ratings100
```

### Задание 4

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе Материалы для лекции «Продвинутый pandas» ---- Ноутбуки к лекции «Продвинутый pandas»):

- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

Необходимо сформировать две таблицы:
* таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
* аналогичную таблицу по типам выручки с указанием адреса клиента

NB! Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
table1 = rzd.merge(auto, how='outer', on='client_id')
table1 = table1.merge(air, how='outer', on='client_id')

table1.loc[table1.rzd_revenue.isnull(), 'rzd_revenue',]=0
table1.loc[table1.auto_revenue.isnull(), 'auto_revenue',]=0
table1.loc[table1.air_revenue.isnull(), 'air_revenue',]=0

table1
```

```python
table2 = table1.merge(client_base, how='outer', on='client_id')
table2
```
